```python
import pandas as pd
import os
import json


path1="E:/srinivas/Phonepe_pulse_data/data/aggregated/transaction/country/india/state/"
agg_trans_list=os.listdir(path1)

col1={'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
        'Transaction_amount': []}

for state in agg_trans_list:
    cur_state=path1+state+"/"
    agg_year_list=os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year=cur_state+year+"/"
        agg_file_list=os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file=cur_year+file
            data=open(cur_file,'r')
            A=json.load(data)
            
            for i in A['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                col1['Transaction_type'].append(name)
                col1['Transaction_count'].append(count)
                col1['Transaction_amount'].append(amount)
                col1['State'].append(state)
                col1['Year'].append(year)
                col1['Quarter'].append(int(file.strip('.json')))
            
df_agg_trans=pd.DataFrame(col1)
            
```

```python
path2="E:/srinivas/Phonepe_pulse_data/data/aggregated/user/country/india/state/"
agg_user_list=os.listdir(path2)

col2={'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [],
            'Percentage': []}

for state in agg_user_list:
    cur_state=path2+state+"/"
    agg_year_list=os.listdir(cur_state)
    
    for year in agg_year_list:
        cur_year=cur_state+year+"/"
        agg_file_list=os.listdir(cur_year)
        
        for file in agg_file_list:
            cur_file=cur_year+file
            data=open(cur_file,'r')
            B=json.load(data)
            try:
                for i in B['data']['usersByDevice']:
                    brand=i['brand']
                    count=i['count']
                    percents=i['percentage']
                    col2['Brands'].append(brand)
                    col2['Count'].append(count)
                    col2['Percentage'].append(percents)
                    col2['State'].append(state)
                    col2['Year'].append(year)
                    col2['Quarter'].append(int(file.strip('.json')))
            except:
                pass

df_agg_user=pd.DataFrame(col2)
```

```python
path3="E:/srinivas/Phonepe_pulse_data/data/map/transaction/hover/country/india/state/"
map_trans_list=os.listdir(path3)

col3={'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [],
            'Amount': []}

for state in map_trans_list:
    cur_state=path3+state+"/"
    map_year_list=os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year=cur_state+year+"/"
        map_file_list=os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file=cur_year+file
            data=open(cur_file,'r')
            C=json.load(data)
            
            for i in C['data']['hoverDataList']:
                dist=i['name']
                count=i['metric'][0]['count']
                amount=i['metric'][0]['amount']
                col3['District'].append(dist)
                col3['Count'].append(count)
                col3['Amount'].append(amount)
                col3['State'].append(state)
                col3['Year'].append(year)
                col3['Quarter'].append(int(file.strip('.json')))
                
df_map_trans=pd.DataFrame(col3)
```

```python
path4="E:/srinivas/Phonepe_pulse_data/data/map/user/hover/country/india/state/"
map_user_list=os.listdir(path4)

col4={"State": [], "Year": [], "Quarter": [], "District": [],
            "RegisteredUser": [], "AppOpens": []}

for state in map_user_list:
    cur_state=path4+state+"/"
    map_year_list=os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year=cur_state+year+"/"
        map_file_list=os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file=cur_year+file
            data=open(cur_file,'r')
            D=json.load(data)
            
            for i in D['data']['hoverData'].items():
                dist=i[0]
                reg_user=i[1]['registeredUsers']
                app=i[1]['appOpens']
                col4['District'].append(dist)
                col4['RegisteredUser'].append(reg_user)
                col4['AppOpens'].append(app)
                col4['State'].append(state)
                col4['Year'].append(year)
                col4['Quarter'].append(int(file.strip('.json')))
                
df_map_user=pd.DataFrame(col4)
```

```python
path5="E:/srinivas/Phonepe_pulse_data/data/top/transaction/country/india/state/"
top_trans_list=os.listdir(path5)

col5={'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [],
            'Transaction_amount': []}

for state in top_trans_list:
    cur_state = path5 + state + "/"
    top_year_list = os.listdir(cur_state)
    
    for year in top_year_list:
        cur_year = cur_state + year + "/"
        top_file_list = os.listdir(cur_year)
        
        for file in top_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            E = json.load(data)
            
            for i in E['data']['pincodes']:
                name = i['entityName']
                count = i['metric']['count']
                amount = i['metric']['amount']
                col5['Pincode'].append(name)
                col5['Transaction_count'].append(count)
                col5['Transaction_amount'].append(amount)
                col5['State'].append(state)
                col5['Year'].append(year)
                col5['Quarter'].append(int(file.strip('.json')))
                
df_top_trans = pd.DataFrame(col5)
```

```python
path6="E:/srinivas/Phonepe_pulse_data/data/top/user/country/india/state/"
top_user_list=os.listdir(path6)

col6={'State': [], 'Year': [], 'Quarter': [], 'Pincode': [],
            'RegisteredUsers': []}

for state in top_user_list:
    cur_state = path6 + state + "/"
    top_year_list = os.listdir(cur_state)
    
    for year in top_year_list:
        cur_year = cur_state + year + "/"
        top_file_list = os.listdir(cur_year)
        
        for file in top_file_list:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            F = json.load(data)
            
            for i in F['data']['pincodes']:
                name = i['name']
                reg_users = i['registeredUsers']
                col6['Pincode'].append(name)
                col6['RegisteredUsers'].append(reg_users)
                col6['State'].append(state)
                col6['Year'].append(year)
                col6['Quarter'].append(int(file.strip('.json')))
df_top_user = pd.DataFrame(col6)
```

```python
mycursor.execute(
    "create table if not exists agg_trans(State varchar(100), Year int, Quarter int, Transaction_type varchar(100), Transaction_count int, Transaction_amount double)")

for i,row in df_agg_trans.iterrows():
    query="insert into agg_trans values(%s,%s,%s,%s,%s,%s)"
    mycursor.execute(query,tuple(row))
    mydb.commit()
```
